```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("spotify_data.csv")
df.head()
```

```python
fig, ax = plt.subplots(figsize=(20,20))
ax.plot(df['Date'],df['Shape of You'],color='red',marker='o')
ax.set_xlabel("Date")
ax.set_ylabel("No of Listeners of Shape of You song")
ax2=ax.twinx()
ax2.plot(df['Date'],df['Despacito'],color='blue',marker='o')
ax2.set_ylabel("No of Listeners of Despacito song")
plt.show()
```

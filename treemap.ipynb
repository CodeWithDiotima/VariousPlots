```python
import numpy as np
import pandas as pd
import squarify as sf
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
dataset=pd.read_csv("IMDB-Movie-Data.csv")
dataset.head()
```

```python
data_by_type=dataset.groupby('Year')['Title'].count()
sizes=[]
labels=[]
for item_type, item_count in data_by_type.items():
  sizes.append(item_count)
  labels.append(item_type)
plt.figure(figsize=(10,6))
plt.title("Tree Map of Year of Release of IMDB Content")
sf.plot(sizes=sizes, label=labels, alpha=0.8)
plt.show()
```
